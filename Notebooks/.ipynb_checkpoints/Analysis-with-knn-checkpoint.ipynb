```python
# Loading libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import warnings
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
warnings.filterwarnings('ignore')
```

```python
data =pd.read_csv('../data/maintenance_servicehours.csv')
display(data.head())
```

```python
data.shape
```

```python
# X/y split
y = data['ServiceHours']
X = data.drop(['ServiceHours'], axis=1)

# Train/Test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 1)

X_train_df = pd.DataFrame(X_train, columns=X.columns)

X_test_df = pd.DataFrame(X_test, columns=X.columns)
```

```python
# data needs to be transformed
transformer = PowerTransformer()
transformer.fit(X_train)

X_train_transformed_np = transformer.transform(X_train)
X_test_transformed_np  = transformer.transform(X_test)

X_train_transformed_df = pd.DataFrame(X_train_transformed_np)
X_test_transformed_df  = pd.DataFrame(X_test_transformed_np)
```

```python
# defining another transformer for y

transformer_y = PowerTransformer()
y_train = np.array(y_train).reshape(-1, 1)

#fitting
y_train_transformed = transformer_y.fit_transform(y_train)
```

```python
import os
import pickle

# Saving transformer
path = "transformers/"
isExist = os.path.exists(path)
if not isExist:
    os.makedirs(path)
    print("The new directory is created!")

filename = "power_transformer_knn.pkl"
with open(path+filename, "wb") as file:
    pickle.dump(transformer, file)

filename = "power_transformer_knn_y.pkl"
with open(path+filename, "wb") as file:
    pickle.dump(transformer_y, file)
```

```python
# Defining the grid of hyperparameters

from sklearn.model_selection import GridSearchCV
grid = {
    'n_neighbors': [2, 3, 4, 5],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]
}

# KNeighborsRegressor model
model = KNeighborsRegressor()

# GridSearchCV
grid_search = GridSearchCV(estimator=model, param_grid=grid, cv=5)

# Fit the model to the data
grid_search.fit(X_train_transformed_df, y_train_transformed)  

# Print the best hyperparameters
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Estimators:", grid_search.best_estimator_)
```

```python
# fittinf the best parameters
best_params = grid_search.best_params_
best_knn = KNeighborsRegressor(**best_params)
best_knn.fit(X_train_transformed_df, y_train_transformed)
```

```python
#Predictions

#y_tpred = best_knn.predict(X_train)
#y_pred = transformer_y.inverse_transform(y_tpred.reshape(-1, 1))

from sklearn.metrics import mean_squared_error, r2_score

y_pred = best_knn.predict(X_test)
# RMSE
rmse = mean_squared_error(y_test, y_pred, squared=False)
print(f'Root Mean Squared Error (RMSE): {rmse:.2f}')

# MSE
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error (MSE): {mse:.2f}')

# R2
r2 = r2_score(y_test, y_pred)
print(f'R-squared: {r2: .2f}')
```

# Plotting model predictions

```python
plt.figure(figsize=(6, 4))
plt.scatter(y_test, y_pred, color='green', alpha=0.5, label='Real vs Predicted')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='fit line')
plt.xlabel("Real Values")
plt.ylabel("Predicted Values")
plt.title("Real vs. Predicted ")
plt.legend()
plt.show()
```

```python

# saving the model

path = "models/"
isExist = os.path.exists(path)
if not isExist:
    os.makedirs(path)
    print("The new directory is created!")

filename = "knn_model.pkl"
with open(path+filename, "wb") as file:
    pickle.dump(best_knn, file)
```
